# 2. Examine the data by using visualizations.


```python
plt.figure(figsize=(8,6))
sns.barplot(x='country',y='International trade: Imports (million US$)',data=df.sort_values('International trade: Imports (million US$)',ascending=False).head(10))
plt.title('Top 10 countries on the basis of International trade:Imports',fontweight='bold',fontsize=15)
plt.xlabel('Country',fontsize=15)
plt.ylabel('International trade: Imports (million US$)',fontsize=15)
plt.xticks(rotation=80)
plt.yticks()
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.barplot(x='country',y='Agricultural production index ',data=df.sort_values('Agricultural production index ',ascending=False).head(10))
plt.title('Top 10 countries based on Agricultural production index',fontweight='bold',fontsize=15)
plt.xlabel('Country',fontsize=15)
plt.ylabel('Agricultural production index',fontsize=15)
plt.xticks(rotation=80)
plt.yticks()
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.scatter(df['Surface area (km2)'],df['Population in thousands (2017)'],s=10,c='red')
plt.title('Surface area VS Population',fontweight='bold',fontsize=18)
plt.xlabel('Surface Area')
plt.ylabel('Population')
```

```python
plt.figure(figsize=(8,4))
plt.scatter(df['Population in thousands (2017)'],df['GDP per capita (current US$)'],s=12,c='green')
plt.title('GDP VS Population',fontweight='bold',fontsize=18)
plt.xlabel('Population in thousands (2017)')
plt.ylabel('GDP per capita (current US$)')
```

# 3. Find out how many countries with respect to the region are available in the dataset


```python
df.groupby("Region")[["country"]].count()
```

# 4. On average which region has the highest


# a. GDP per capita (current US$)

```python
gdp=df.groupby("Region")[["GDP per capita (current US$)"]].mean()
a=gdp.sort_values(by="GDP per capita (current US$)",ascending=False)
a
```

```python
a.nlargest(1,"GDP per capita (current US$)")
```

# b. International trade: Imports (million USD


```python
impor=df.groupby("Region")[["International trade: Imports (million US$)"]].mean()
n=impor.sort_values(by="International trade: Imports (million US$)",ascending=False)
n
```

```python
n.nlargest(1,"International trade: Imports (million US$)")
```

# 5. Find out which region has more consistent with respect to GDP per capita (current USD

```python
consistent=df.groupby("Region")[["GDP per capita (current US$)"]].std()
consistent.nsmallest(1,"GDP per capita (current US$)")
```
