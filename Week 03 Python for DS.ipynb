# Find the minimum and maximum value of insulin level in the data.

```python
print("minimum value of insulin level =",data6['Insulin'].min())
print("maximum value of insulin level =",data6['Insulin'].max())
```

```python
#Create a histogram of "Age".
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
plt.hist(data6['Age'],color='r',rwidth=0.9)
plt.title('Histogram plot of Age',fontsize=19)
plt.xlabel('Age',fontsize=16)
plt.ylabel('Frequency',fontsize=16)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
```

# Create a scatter plot of "BMI" vs "Glucose".

```python
plt.figure(figsize=(8,8))
plt.scatter(data6['BMI'],data6['Glucose'],c='c',s=20)
plt.title('plot of BMI vs Glucose',fontsize=19)
plt.xlabel('BMI')
plt.ylabel('Glucose')
```

# Replace the min value of "BMI" with mean value of BMI

```python
import numpy as np
minimum=data6['BMI'].min()
print("minimum= ",minimum)
mean=np.mean(data6['BMI'])
print("mean= ",mean)
```

```python
data6.loc[(data6["BMI"]==0),"BMI"]=data6["BMI"].mean()
data6['BMI'].min()
```

# Draw a grouped boxplot with "Outcome" on x-axis and "Glucose" on the y-axis such that we see distributions of Glucose separately for each outcome.

```python
import seaborn as sns
plt.figure(figsize=(8,8))
sns.boxplot(x='Outcome',y='Glucose',data=data6)
plt.title("Boxplot of outcome vs glucose ",fontsize=19)
```

### 5.

**-** Create a numpy array (first) with 5 x 5 size and all elements are 0 in that.

**-** Create a numpy array (second) with 5 x 5 and all elements are 1 in that.

**-** Combine the arrays first and second vertically, the final array size should be 10 x 5

```python
import numpy as np
arr1=np.zeros((5,5),dtype=int)
print(arr1)
```

```python
arr2=np.ones((5,5),dtype=int)
print(arr2)
```

```python
arr3=np.concatenate((arr1,arr2),axis=0)
print(arr3)
```
